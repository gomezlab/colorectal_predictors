```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams.update({'font.size': 16})
```

```python
%store -r rf_fpr_ssi
%store -r rf_tpr_ssi
%store -r xgb_fpr_ssi
%store -r xgb_tpr_ssi

%store -r lr_fpr_ssi
%store -r lr_tpr_ssi
%store -r ann_fpr_ssi
%store -r ann_tpr_ssi
```

```python
plt.figure(num=None, figsize=(10, 9), dpi=300, facecolor='w', edgecolor='k')

plt.plot(rf_fpr_ssi, rf_tpr_ssi, color='royalblue', lw=2, alpha=0.8, label='Random Forest (0.764)')
plt.plot(xgb_fpr_ssi, xgb_tpr_ssi, color='lightsteelblue', lw=2, alpha=0.8,  label='Gradient Boosting (0.766)')
plt.plot(ann_fpr_ssi, ann_tpr_ssi, color='cornflowerblue', lw=2, alpha=0.8,  label='Neural Network (0.770)')
plt.plot(lr_fpr_ssi, lr_tpr_ssi, color='red', lw=2, alpha=0.8,  label='Logistic Regression (0.677)')
plt.legend(loc="lower right")
plt.xlabel('1 - Specificity')
plt.ylabel('Sensitivity')
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.savefig('roc.png', bbox_inches='tight')
```

```python
%store -r rf_prec_ssi
%store -r rf_rec_ssi
%store -r xgb_prec_ssi
%store -r xgb_rec_ssi
%store -r lr_prec_ssi
%store -r lr_rec_ssi
%store -r ann_prec_ssi
%store -r ann_rec_ssi
```

```python
plt.figure(num=None, figsize=(10, 9), dpi=300, facecolor='w', edgecolor='k')
plt.plot(rf_rec_ssi, rf_prec_ssi, color='royalblue', lw=2, alpha=0.8, label='Random Forest (0.484)')
plt.plot(xgb_rec_ssi, xgb_prec_ssi, color='lightsteelblue', lw=2, alpha=0.8, label='Gradient Boosting (0.483)')
plt.plot(ann_rec_ssi, ann_prec_ssi, color='cornflowerblue', lw=2, alpha=0.8, label='Neural Network (0.488')
plt.plot(lr_rec_ssi, lr_prec_ssi, color='red', lw=2, alpha=0.8, label='Logistic Regression (0.194)')
plt.legend(loc="upper right")
plt.axis([0, 1, 0, 1])                                    
plt.xlabel('Recall (Sensitivity)', fontsize=16) 
plt.ylabel('Precision (PPV)', fontsize=16)                                                                               
plt.savefig('prc.png', bbox_inches='tight')
plt.show()
```
