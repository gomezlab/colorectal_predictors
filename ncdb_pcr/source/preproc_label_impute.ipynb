```python
cat_imputer = SimpleImputer(strategy='constant', fill_value='unknown')
cols = data.columns.to_list()
# num_cols = ['Age at Diagnosis', 'Great Circle Distance', 'Size of Tumor', 'Surgical Dx and Staging Procedure, Days from Dx','CEA value', 'Treatment Started, Days from Dx', 'First Surgical Procedure, Days from Dx','Definitive Surgical Procedure, Days from Dx','Radiation, Days from Dx','Radiation Ended, Days from Start of Radiation','Systemic, Days from Dx','Chemotherapy, Days from Dx', 'Total Dose','days_chemo_rads']
num_cols = ['Age at Diagnosis', 'Great Circle Distance', 'Size of Tumor', 'CEA value']
cat_cols = [x for x in cols if x not in num_cols]
```

```python
#set datatype of categorical columns to object and fill with 'unknown'
for col in cat_cols:
    data[col] = data[col].astype(str)
for x in cat_cols:
    data[x] = cat_imputer.fit_transform(data[x].values.reshape(-1,1))
```

```python
#fill numerical columns with median
num_imputer = SimpleImputer(strategy='median')
for x in num_cols:
    data[x] = num_imputer.fit_transform(data[x].values.reshape(-1,1))
```

```python
#label encode the categorical columns
data.head()
cat_cols.remove('Year of Diagnosis')
for x in cat_cols:
    le = LabelEncoder()
    le.fit(data[x])
    keys = le.classes_
    values = le.transform(le.classes_)
    label_dict = dict(zip(keys, values))
    print('col:', x)
    print(label_dict)
    data[x] = le.transform(data[x])
```

```python
#apply the standard scaler to the numerical columns
scaler = RobustScaler()
for x in num_cols:
    data[x] = scaler.fit_transform(data[x].values.reshape(-1,1))
```
