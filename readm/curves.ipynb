```python
from matplotlib import pyplot as plt
```

```python
import matplotlib
```

```python
matplotlib.rcParams.update({'font.size': 16})
```

```python
%store -r rf_fpr_readm
%store -r rf_tpr_readm
%store -r xgb_fpr_readm
%store -r xgb_tpr_readm
```

```python
%store -r lr_fpr_readm
%store -r lr_tpr_readm
%store -r ann_fpr_readm
%store -r ann_tpr_readm
```

```python
plt.figure(num=None, figsize=(10, 10), dpi=300, facecolor='w', edgecolor='k')
plt.plot(rf_fpr_readm, rf_tpr_readm, color='royalblue', lw=3, label='Random Forest (AUC = 0.748)')
plt.plot(xgb_fpr_readm, xgb_tpr_readm, color='lightsteelblue', lw=3, label='Gradient Boosting (AUC = 0.745)')
plt.plot(ann_fpr_readm, ann_tpr_readm, color='cornflowerblue', lw=3, label='Neural Network (AUC = 0.751)')
plt.plot(lr_fpr_readm, lr_tpr_readm, color='red', lw=3, label='Logistic Regression (AUC = 0.684)')
plt.legend(loc="lower right")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.savefig('roc.png', bbox_inches='tight', dpi=600)
```

```python
%store -r rf_prec_readm
%store -r rf_rec_readm
%store -r xgb_prec_readm
%store -r xgb_rec_readm
%store -r lr_prec_readm
%store -r lr_rec_readm
%store -r ann_prec_readm
%store -r ann_rec_readm
```

```python
plt.figure(num=None, figsize=(10, 10), dpi=300, facecolor='w', edgecolor='k')
plt.plot(rf_rec_readm, rf_prec_readm, color='royalblue', lw=3, label='Random Forest (AUC = 0.378)')
plt.plot(xgb_rec_readm, xgb_prec_readm, color='lightsteelblue', lw=3, label='Gradient Boosting (AUC = 0.384)')
plt.plot(ann_rec_readm, ann_prec_readm, color='cornflowerblue', lw=3, label='Neural Network (AUC = 0.405)')
plt.plot(lr_rec_readm, lr_prec_readm, color='red', lw=3, label='Logistic Regression (AUC = 0.194)')
plt.legend(loc="upper right")
plt.axis([0, 1, 0, 1])
plt.xlabel('Recall (Sensitivity)', fontsize=16)
plt.ylabel('Precision (PPV)', fontsize=16)
plt.savefig('prc.png', bbox_inches='tight', dpi=600)
```
